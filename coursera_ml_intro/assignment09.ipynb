```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('../data/classification.csv')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
acc = accuracy_score(df['true'], df['pred'])
prec = precision_score(df['true'], df['pred'])
recall = recall_score(df['true'], df['pred'])
f1 = f1_score(df['true'], df['pred'])
print(f'Task 2 : {acc} {prec} {recall} {f1}')
```

```python
# TP, FP, FN Ð¸ TN
TP = df[(df['true'] == 1) & (df['pred'] == 1)]
FP = df[(df['true'] == 0) & (df['pred'] == 1)]
FN = df[(df['true'] == 1) & (df['pred'] == 0)]
TN = df[(df['true'] == 0) & (df['pred'] == 0)]
print(f'Task 1 : {len(TP)} {len(FP)} {len(FN)} {len(TN)}')
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
data = []
for col_name in ('score_logreg', 'score_svm', 'score_knn', 'score_tree'):
    prec, recall, thresholds = precision_recall_curve(df2['true'], df2[col_name])
    trace = go.Scatter(x=recall, y=prec, name=col_name)
    data.append(trace)

layout = {'title': 'Precision Recall Curves'}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
