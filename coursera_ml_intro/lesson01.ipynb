```python
X = np.random.normal(loc=1, scale=10, size=(1000, 50))
```

```python
X_dev = np.std(X, axis=0)
```

```python
X_mean, X_dev
```

```python
np.mean(X - X_mean)
```

```python
X_norm = (X - X_mean) / X_dev
```
