```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('../data/svm-data.csv', names=['Target', 'X1', 'X2'], header=None)
```

```python
dataset
```

```python
dataset['Target'] = dataset['Target'].astype('int64')
```

```python
dataset
```

```python
y = dataset['Target']
X = dataset[['X1', 'X2']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'svg'
```

```python
sns.lmplot('X1', 'X2', data=dataset, hue='Target', fit_reg=False)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear', C=100000, random_state=241)
```

```python
svc.fit(X, y)
```
